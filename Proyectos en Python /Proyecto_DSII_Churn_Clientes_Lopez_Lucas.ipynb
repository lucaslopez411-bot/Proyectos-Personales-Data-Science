#### Contexto:
1. **Introducción: Retención del cliente**

La sostenibilidad del negocio bancario depende en gran medida de su capacidad para retener clientes y ofrecer servicios que respondan a sus necesidades. La pérdida de clientes (churn) no solo implica una disminución directa de ingresos, sino también mayores costos de adquisición para reemplazarlos. En un entorno altamente competitivo, anticipar el abandono se vuelve un factor estratégico clave.

2. **Planteamiento del Problema: Detección Reactiva del Churn**

Los sistemas tradicionales de gestión de clientes suelen actuar de forma reactiva, identificando la pérdida cuando el cliente ya ha tomado la decisión de abandonar la entidad. Este enfoque limita la capacidad de intervención temprana y reduce la efectividad de las estrategias de fidelización.

3. **Justificación: El Valor Predictivo de los Datos**

El Machine Learning permite migrar hacia un enfoque proactivo mediante el análisis de patrones de comportamiento financiero y relacional. A partir de variables transaccionales, demográficas y de uso de productos, es posible construir modelos predictivos que funcionen como sistemas de alerta temprana para identificar clientes con alta probabilidad de abandono.

# Entrenamiento Supervisado de Clasificación

En el siguiente paso, comenzamos con la preparación y división de datos para nuestro modelo de ML de clasificación, donde separamos los datos en conjuntos de entrenamiento y prueba. En esta ocasión se han destinado 75% de los datos para entrenamiento y un 25% para prueba, que nos permitiran predecir la variable de interés "Churn".

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

A continuación con nuestros datos entrenados vamos a aplicar 3 modelos distintos para corroborar cual se adapta mejor a nuestro conjunto de datos. Para ello vamos a usar "Logistic Regression" (Regresión Logística), "Random Forest" (Arboles de decisión) y XGBoost. A su vez, para entrenar los modelos lo haremos con un balanceo entre las clases.

```python
scale_pos_weight = (y_train == 0).sum() / (y_train == 1).sum()

models = {'Logistic Regression': LogisticRegression(random_state=42, max_iter=1000, class_weight='balanced'),
          'Random Forest': RandomForestClassifier(random_state=42, n_estimators=100, class_weight='balanced'),
          'XGBoost': XGBClassifier(random_state=42, eval_metric='logloss')}
```

Ahora definimos una función que entrena un modelo de clasificación con los datos de entrenamiento y evalúa su desempeño sobre el conjunto de prueba. Calcula métricas clave como accuracy, precision, recall, F1-score y AUC, además de realizar validación cruzada con StratifiedKFold para obtener una medida más robusta del rendimiento. Finalmente, devuelve las métricas junto con las predicciones y probabilidades del modelo.

```python
def evaluar_modelo_clasificacion(model, X_train, X_test, y_train, y_test):

    model.fit(X_train, y_train)

    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1]

    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    auc = roc_auc_score(y_test, y_pred_proba)

    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
    cv_scores = cross_val_score(model, X_train, y_train, cv=cv, scoring='roc_auc')

    return {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1-Score': f1,
        'AUC': auc,
        'CV Mean': cv_scores.mean(),
        'CV Std': cv_scores.std()
    }, y_pred, y_pred_proba
```

```python
print("ENTRENANDO Y EVALUANDO MODELOS")

results = {}
predictions = {}
y_pred_proba_dict = {}

for name, model in models.items():
    print(f"\n Entrenando {name}...")

    if name in ['Logistic Regression']:
        print(f"Generando data scaled para : {name}")
        X_tr, X_te = X_train_scaled, X_test_scaled
    else:
        print(f"Generando data para : {name}")
        X_tr, X_te = X_train, X_test

    metrics, y_pred, y_pred_proba = evaluar_modelo_clasificacion(
        model, X_tr, X_te, y_train, y_test
    )

    results[name] = metrics
    predictions[name] = y_pred
    y_pred_proba_dict[name] = y_pred_proba

    print(f"{name} completado - Accuracy: {metrics['Accuracy']:.3f}")

print(" RESULTADOS COMPARATIVOS ")

results_df = pd.DataFrame(results).T
results_df = results_df.round(4)
results_df_sorted = results_df.sort_values('AUC', ascending=False)
results_df_sorted
```

Se puede visualizar a través de la tabla que el modelo que mejor se adapta a nuestro conjunto de datos es XGBoost, ya que la mayoría de las métricas son mejores en dicho modelo. Especialmente AUC porque gestiona eficazmente los conjuntos de datos desequilibrados y proporciona una medida de rendimiento independiente del umbral de decisión elegido, evalúa la capacidad del modelo para clasificar correctamente tanto a los que se van como a los que se quedan, sin dejarse engañar por la clase mayoritaria. Vamos a validar a continuación la matriz de confusión de XGBoost.

```python
best_model_clasif = results_df_sorted.index[0]
print(f"Mejor modelo: {best_model_clasif}")

y_pred_best = predictions[best_model_clasif]

print("\nClassification Report:")
print(classification_report(y_test, y_pred_best))

plt.figure(figsize=(4,3))
sns.heatmap(confusion_matrix(y_test, y_pred_best),
            annot=True, fmt='d', cmap='Reds')
plt.title(f'Matriz de Confusión - {best_model_clasif}')
plt.xlabel('Predicho')
plt.ylabel('Real')
plt.show()
```

El modelo XGBoost presenta un rendimiento general sólido, con una exactitud global del 90%.

En cuanto a la clase minoritaria (1), correspondiente a los clientes que abandonan, el modelo alcanza un recall de 0.63, lo que implica que logra detectar el 63% de los casos reales de churn. En conjunto, el modelo logra un equilibrio efectivo entre la identificación de clientes en riesgo y la precisión de las predicciones, consolidándose como una herramienta valiosa para optimizar las estrategias de retención y priorizar esfuerzos hacia los clientes con mayor probabilidad de abandono.

A continuación, realizaremos una optimización de este modelo con Grid Search CV, para verificar si hay posibles mejoras sobre el mismo y comparandolo luego con el modelo de Random Forest, tanto para clasificación como Regresión (es decir, se utilizará para comparar el mejor modelo con Random Forest de ser necesario). Para ello definimos la siguiente función:

```python
def tuning_modelo(model, param_grid, X_train, y_train, tipo='clasificacion'):

    print(f"\nTuning: {model.__class__.__name__}")

    if tipo == 'clasificacion':
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
        scoring = 'roc_auc'
    else:
        cv = KFold(n_splits=5, shuffle=True, random_state=42)
        scoring = 'r2'

    grid = GridSearchCV(
        estimator=model,
        param_grid=param_grid,
        scoring=scoring,
        cv=cv,
        n_jobs=-1,
        verbose=1
    )

    grid.fit(X_train, y_train)

    print("Best params:", grid.best_params_)
    print("Best score:", grid.best_score_)

    return grid.best_estimator_
```

Ahora definimos los hiperparámetros a optimizar para Random Forest y XGBoost y luego hacemos recorrer estos modelos para aplicar GridSearchCV mediante la función "tuning_modelo".
En cada iteración, entrena el modelo con distintas combinaciones de parámetros y guarda la mejor versión optimizada en el diccionario best_models_clasif.

```python
param_grids_clasif = {
    'Random Forest': {
        'n_estimators': [50, 100],
        'max_depth': [3, 5],
        'min_samples_split': [2, 5]
    },
    'XGBoost': {
        'n_estimators': [50, 100],
        'max_depth': [3, 5],
        'learning_rate': [0.05, 0.1]
    }
}

best_models_clasif = {}

for name in ['Random Forest', 'XGBoost']:

    model = models[name]

    if name == 'Logistic Regression':
        X_tr = X_train_scaled
    else:
        X_tr = X_train

    best_model = tuning_modelo(
        model,
        param_grids_clasif[name],
        X_tr,
        y_train,
        tipo='clasificacion'
    )

    best_models_clasif[name] = best_model
```

Ahora realizamos una función de comparación entre los dos modelos de clasificación sobre la metrica AUC e iteramos para almacenar los datos en ua nueva variable llamada "performance_clasif".

```python
def comparar_clasificacion(model_base, model_tuned, X_train, X_test, y_train, y_test):

    # Entrenamiento de ambos modelos
    model_base.fit(X_train, y_train)
    model_tuned.fit(X_train, y_train)

    # Probabilidades de ambos modelos
    y_base_proba = model_base.predict_proba(X_test)[:, 1]
    y_tuned_proba = model_tuned.predict_proba(X_test)[:, 1]

    # Métricas de ambos modelos
    auc_base = roc_auc_score(y_test, y_base_proba)
    auc_tuned = roc_auc_score(y_test, y_tuned_proba)

    return auc_base, auc_tuned
```

```python
performance_clasif = {}

for name in ['Random Forest', 'XGBoost']:

    model_base = models[name]
    model_tuned = best_models_clasif[name]

    X_tr = X_train
    X_te = X_test

    auc_base, auc_tuned = comparar_clasificacion(
        model_base, model_tuned,
        X_tr, X_te,
        y_train, y_test
    )

    performance_clasif[name] = [auc_base, auc_tuned]

performance_clasif
```

Ahora definimos una función que entrena un modelo de regresión con los datos de entrenamiento y evalúa su desempeño sobre el conjunto de prueba. Calcula métricas clave como MAE, RMSE, R2, CV Mean R2 y CV std, además de realizar validación cruzada para obtener una medida más robusta del rendimiento. Finalmente, devuelve las métricas junto con las predicciones y probabilidades del modelo. La misma se va a utilizar para los modelos de Linear Regression (Regresión Lineal), Random Forest Regressor (Arbol de Decisión de Regresión) y Gradient Boosting Regressor.

```python
def evaluar_modelo_regresion(model, X_train, X_test, y_train, y_test, X_full, y_full):

    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    mae = mean_absolute_error(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2 = r2_score(y_test, y_pred)

    cv_scores = cross_val_score(model, X_full, y_full, cv=3, scoring='r2')

    return {
        'MAE': mae,
        'RMSE': rmse,
        'R2': r2,
        'CV Mean R2': cv_scores.mean(),
        'CV Std': cv_scores.std()
    }
```

```python
features_cluster = [
    'Credit_Limit',
    'Total_Trans_Amt',
    'Total_Trans_Ct',
    'Avg_Utilization_Ratio'
]

X_unsup = df_corr[features_cluster]
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_unsup)
```

El siguiente paso es graficar el Método del Codo (Elbow Method) para una primera aproximación a identificar el tamaño de Clusters ideales.

```python
inertia = []
K = range(2, 10)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(6,4))
plt.plot(K, inertia, marker='o')
plt.xlabel('Número de clusters (k)')
plt.ylabel('Inercia')
plt.title('Método del Codo')
plt.show()
```

Se puede observar en este caso, que el número ideal de clusters sería 5, para corroborar esto, utilizaremos el índice de Silhouette (Silhouette Score) tanto numéricamente como gráficamente, de esta manera sabremos si este es el número ideal de clusters.

```python
for k in range(2, 10):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    print(f"k={k} → Silhouette Score: {score:.4f}")
```

```python
range_n_clusters = [2,3,4,5,6,7,8,9]

for n_clusters in range_n_clusters:

    fig, ax1 = plt.subplots(figsize=(8, 5))

    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(X_scaled)

    silhouette_avg = silhouette_score(X_scaled, cluster_labels)
    print(f"Para n_clusters = {n_clusters}, Silhouette Score promedio = {silhouette_avg:.4f}")

    sample_silhouette_values = silhouette_samples(X_scaled, cluster_labels)

    y_lower = 10

    for i in range(n_clusters):
        ith_cluster_silhouette_values = sample_silhouette_values[cluster_labels == i]
        ith_cluster_silhouette_values.sort()

        size_cluster_i = ith_cluster_silhouette_values.shape[0]
        y_upper = y_lower + size_cluster_i

        color = cm.nipy_spectral(float(i) / n_clusters)

        ax1.fill_betweenx(
            np.arange(y_lower, y_upper),
            0,
            ith_cluster_silhouette_values,
            facecolor=color,
            edgecolor=color,
            alpha=0.7
        )

        ax1.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))

        y_lower = y_upper + 10

    ax1.set_title(f"Gráfico de Silhouette para k = {n_clusters}")
    ax1.set_xlabel("Coeficiente de Silhouette")
    ax1.set_ylabel("Clusters")

    ax1.axvline(x=silhouette_avg, color="red", linestyle="--")

    ax1.set_yticks([])
    ax1.set_xlim([-0.1, 1])

    plt.show()
```

Tras analizar el coeficiente de Silhouette y la distribución visual de los grupos para K=, se ha determinado que K=6 es el número óptimo de clusters para la segmentación de la base de clientes.

Esta decisión se fundamenta en dos pilares:

1. Estabilidad y Rendimiento: El modelo con K=6 alcanza un coeficiente de Silhouette de 0.3893, un valor altamente competitivo y muy cercano al máximo obtenido en las pruebas. Este puntaje confirma una coherencia interna sólida y una separación clara entre los grupos.
2. Equilibrio y Calidad de la Segmentación (Evitación de la sobre-segmentación): Al comparar visualmente el gráfico de K=6 frente al de K=7 y K=2, observamos que en ambos los tamaños de los clusters son significativamente desproporcionados (se observa un grupo extremadamente reducido). Esto sugiere un fenómeno de "sobre-segmentación", donde el modelo intenta forzar la división de grupos que en realidad comparten características similares. En contraste, K=6 presenta una distribución más equilibrada y sustancial, lo que garantiza que cada segmento tenga el tamaño suficiente para ser representativo y accionable.

A continuación validaremos con Grid Search, KFold y Bagging, que resultados obtenemos.

```python
print("GRID SEARCH PARA KMEANS")

param_grid = {
    'n_clusters': [4, 5, 6, 7],
    'n_init': [10, 20],
    'max_iter': [300, 500]
}

best_score = -1
best_params = None

for params in ParameterGrid(param_grid):

    kmeans = KMeans(
        n_clusters=params['n_clusters'],
        n_init=params['n_init'],
        max_iter=params['max_iter'],
        random_state=42
    )

    labels = kmeans.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)

    if score > best_score:
        best_score = score
        best_params = params

print("Mejores parámetros:", best_params)
print("Mejor Silhouette:", round(best_score, 4))
```

```python
print("VALIDACIÓN CRUZADA (K-FOLD)")

kf = KFold(n_splits=5, shuffle=True, random_state=42)

scores_kf = []

for train_idx, test_idx in kf.split(X_scaled):

    X_fold = X_scaled[train_idx]

    kmeans = KMeans(n_clusters=6, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X_fold)

    score = silhouette_score(X_fold, labels)
    scores_kf.append(score)

print(f"Silhouette promedio: {np.mean(scores_kf):.4f}")
```

```python
print("ANÁLISIS CON BAGGING")

n_runs = 10
silhouette_scores = []

for i in range(n_runs):

    X_sample = resample(X_scaled, replace=True, random_state=i)

    kmeans = KMeans(n_clusters=6, random_state=i, n_init=10)
    labels = kmeans.fit_predict(X_sample)

    score = silhouette_score(X_sample, labels)
    silhouette_scores.append(score)

print(f"Silhouette promedio: {np.mean(silhouette_scores):.4f}")
print(f"Desvío estándar: {np.std(silhouette_scores):.4f}")
```

Con la optimización a través de Grid Search identificamos como mejor configuración un modelo con 7 clusters, alcanzando un Silhouette Score de 0,391. Sin embargo, al aplicar validación cruzada con K-Fold y un enfoque de estabilidad tipo bagging, los resultados muestran valores muy similares (entre 0.389 y 0.390) y con una variabilidad extremadamente baja (desvío estándar de 0,0019).

Esto indica que el modelo es bastante estable en sí y muestra robustez, ya que pequeñas variaciones en los datos no afectan significativamente la calidad en el clustering. Además, la diferencia entre usar 6 o 7 clusters es mínima en comparación de las performances, por lo que la elección final puede apoyarse también en criterios de interpretabilidad y negocio.

En definitiva, la segmentación obtenida es sólida y confiable, validando la calidad del modelo más allá de una única métrica puntual.

### PCA

De esta forma, ya que tenemos seleccionado nuestro número óptimo de clusters, volveremos a usar KMeans en base a este número K que hemos elegido y, a continuación, utilizaremos PCA (Análisis de Componentes Principales) para reducir la dimensionalidad de nuestro dataset original y proyectar los datos en un espacio bidimensional.

Este procedimiento no altera la lógica de agrupación del algoritmo, sino que nos permite visualizar y validar la separación efectiva de dichos grupos. Al comprimir la varianza de las variables originales en los dos componentes principales (CP1 y CP2), podemos verificar gráficamente si los perfiles de clientes identificados por KMeans presentan límites claros o si existen zonas de solapamiento, asegurando así la robustez de la segmentación realizada.

```python
kmeans = KMeans(n_clusters=6, random_state=42)

df['cluster'] = kmeans.fit_predict(X_scaled)
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

plt.figure(figsize=(7,5))

scatter = plt.scatter(
    X_pca[:, 0],
    X_pca[:, 1],
    c=df['cluster'],
    cmap='tab10',
    alpha=0.6)

plt.title("Clusters visualizados con PCA (k=6)")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")

plt.colorbar(scatter)
plt.show()
```

```python
print("Varianza explicada por cada componente:")
print(pca.explained_variance_ratio_)
```

### Conclusión y análisis de la segmentación mediante PCA y Varianza Explicada

Al analizar el gráfico de dispersión PCA con K=6, observamos que los grupos presentan una separación geométrica bastante definida en el plano bidimensional. Para validar la fidelidad de esta representación, hemos calculado la varianza explicada por los dos componentes principales:

Varianza Explicada:

* Componente Principal 1: 47.02%
* Componente Principal 2: 35.53%
* Varianza Total Acumulada: 82.55%

Esta métrica es fundamental: nos indica que el gráfico visualizado retiene el 82.55% de la información original del dataset. En términos de análisis de datos, este es un valor muy robusto. Significa que la estructura de los clusters que observamos en la imagen es una representación altamente confiable de cómo están organizados los datos en su espacio multidimensional original, perdiendo menos del 18% de la información total en el proceso de reducción.

A partir de la visualización y este dato técnico, podemos concluir:

* Representación Fiel: Dado que superamos el umbral del 80% de varianza explicada, podemos afirmar con alta confianza que las distancias y separaciones que vemos entre los grupos (especialmente entre los grupos de tonos azul, verde y rosa) son reales y no un artefacto estadístico de la reducción de dimensiones.
* Interpretación del Solapamiento: El ligero solapamiento que observamos en la zona central (donde convergen el celeste y el verde) es comprensible, pero dado que retenemos el 82.55% de la información, podemos concluir que este solapamiento refleja una similitud real de comportamiento entre esos perfiles en el espacio original, y no una falla de la visualización.
* Robustez: La combinación de una separación visual clara y una varianza explicada superior al 80% valida la configuración de K=6 como una opción óptima. La segmentación resultante es sólida y operativa, ideal para el desarrollo de estrategias de negocio diferenciadas.
